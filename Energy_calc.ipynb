```python
from matplotlib import pyplot as plt
import os
import math
import numpy as np

%matplotlib inline

energy_data=np.loadtxt(os.path.expanduser('~/phys6810/Final_Project/energy_output.dat'),skiprows=2)
concentration=np.loadtxt(os.path.expanduser('~/phys6810/Final_Project/energy_output.dat'),skiprows=1,max_rows=1)
```

the below code take the energy output and plot the energy respect to the number of flip

```python
# set up the plot
fig=plt.figure()
ax1=fig.add_subplot(111)

# axes labels and title
ax1.set_ylabel('t')
ax1.set_xlabel('E(t)')
ax1.set_title('Energy vs time for different concentration of Disorder')

# axes ranges
#ax1.set_xlim([-10,-1])
#ax1.set_ylim([-12,0])

# the actual plot of the data (column 2 vs column 1, but zero-based)

ax1.plot(energy_data[:,0],energy_data[:,1],c='r',label='C=0')
ax1.plot(energy_data[:,0],energy_data[:,2],c='blue',label='C=0.005')
ax1.plot(energy_data[:,0],energy_data[:,3],c='green',label='C=0.01')
ax1.plot(energy_data[:,0],energy_data[:,4],c='pink',label='C=0.03')
ax1.plot(energy_data[:,0],energy_data[:,5],c='cyan',label='C=0.1')
ax1.plot(energy_data[:,0],energy_data[:,6],c='yellow',label='C=0.2')
ax1.plot(energy_data[:,0],energy_data[:,7],c='black',label='C=0.4')

# show the legend
ax1.legend(loc='upper right')

# make sure the plot show up
plt.show()
```

```python
fig.savefig(os.path.expanduser('~/phys6810/Final_Project/E_overflip.jpg'))
```

The below code compare the case for the relationship betweeen DE with concentration of disorder and fit it using log scale

```python


y_temp=energy_data[-1,:]
y=y_temp[2:]-y_temp[1]
x=concentration[1:]
fit_coefficients=np.polyfit(np.log10(x),np.log10(y),deg=1)
fit_title='{:.1f}*x{:+.1f}'.format(fit_coefficients[0],fit_coefficients[1])


# set up the plot
fig=plt.figure()
ax1=fig.add_subplot(111)

# axes labels and title
ax1.set_ylabel('Equlibrium delta E respect to nondoping(log scale)')
ax1.set_xlabel('concentration of disorder(log scale)')
ax1.set_title('Equlibrium Energy vs concentration of Disorder')

# axes ranges
#ax1.set_xlim([-10,-1])
#ax1.set_ylim([-12,0])

# the actual plot of the data (column 2 vs column 1, but zero-based)

ax1.scatter(np.log10(x),np.log10(y),c='cyan')

x=np.linspace(-2.3,-0.5,50)
ax1.plot(x, fit_coefficients[0]*x+fit_coefficients[1],c='b',label=fit_title)

# show the legend
ax1.legend(loc='upper right')

# make sure the plot show up
plt.show()
```
